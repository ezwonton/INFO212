```python
# Write your code below

import pandas as pd
# importing libraries

df = pd.read_csv("loan-applications.csv")
# reading file

df
#printing dataframe
```

```python
# 1.) List the number of missing values for each column.
df.isnull().sum()
```

```python
# 2.) Remove the applications each of which has more than 2 missing values.
df = df.dropna(thresh = 2)
df
```

```python
# 3.) Forward fill up the missing values for columns with Object data type.
for i, t in enumerate(df.dtypes): # for loop over dataframe column types
    if str(t) == "object": # if conditional on columns being the "object" datatype
        df.iloc[:, i].fillna(method = "ffill", inplace = True) # forward filling the column
df
```

```python
# 4.) Fill up the missing values with their means for columns with numerical data types.
for i, t in enumerate(df.dtypes): # for loop over column types
    if str(t) == "int64" or str(t) == "float64": # if conditional on numerical columns
        df.iloc[:, i].fillna(df.iloc[:,i].mean(), inplace = True) # forward filling on numerical columns
df
```

```python
# 5.) After the above steps, show that there are no missing data in the DataFrame.
df.isnull().sum()
```

```python
import pandas as pd
# importing libraries

df = pd.read_csv("imdb_1000.csv") # reading file
df
```

```python
df["actors_list"].head() # getting the first 5 rows to see how the column "actors list" is formatted
```

```python
actorslist = df["actors_list"] # getting column of the imported dataframe
actorslist[147] # noticed some interesting formated rows and wanted to print it out
```

```python
actors1 = actorslist[1].split("'") # testing string splitting on row 1
actors1 # printing out the output
```

```python
actors_split = pd.DataFrame(columns=['Actor1','Actor2','Actor3'])
# new dataframe of 3 actors names

for i, l in enumerate(actorslist): #looping over dataframe rows
    actors = actorslist.iloc[i].split(",")
    # splitting row on ","
    
    actor1 = actors[0].split("u", 1)[1]
    # geting row split index #0
    # splitting that on "u" one time
    # setting actor1 as index 1 of the second splitting
    
    actor2 = actors[1].split("u", 1)[1]
    # geting row split index #1
    # splitting that on "u" one time
    # setting actor1 as index 1 of the second splitting
    
    actor3 = actors[2].split("u", 1)[1][:-1]
    # geting row split index #2
    # splitting that on "u" one time
    # setting actor1 as index 1 ignoring last index of the string of the second splitting
    
    actors_split = actors_split.append({'Actor1':actor1, 'Actor2':actor2, 'Actor3':actor3}, ignore_index = True)
    # appending actors names into appropriate columns
    
actors_split # printing dataframe
```

```python
df = df.merge(actors_split, left_index=True, right_index= True)
# merging two dataframes

df.drop('actors_list', axis=1, inplace=True)
# dropping the original actors list column

df
```
