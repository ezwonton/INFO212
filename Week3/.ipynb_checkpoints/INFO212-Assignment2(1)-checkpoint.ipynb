#### Question 2:
Load the data in the file 'inflammation-05.csv' to an Numpy array. Answer the following questions:
* Suppose the rows represent patients, and columns represent days. How many patients and how many days does the file contain inflammation data for?
* Extract the values for the last 10 patients' inflammation in last 10 days.
* Compute an array containing the average inflammation for all patients across all days. Plot the averages and label the plot with "Average Inflammation of Patients". 
* Compute an array containing the average inflammtion for all days across all patients. Plot the averages and label the plot with "Average Inflammation of Days" 

````python
# Write your code below
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt("inflammation-data/inflammation-05.csv", delimiter = ",")
patients = data.shape[0]
days = data.shape[1]
print("The number of patients is %.f patients, and the number of days is %.f days." %(patients, days))
for i, patient in enumerate(data[-10:]):
    print("Patient #%.f with information %s." %(i+51, patient))

patient_avg = np.mean(data, axis = 1)
plt.plot(patient_avg)
plt.title('Average Inflammation of Patients')
plt.xlabel('Patient')
plt.ylabel('Average')
plt.show()
day_avg = np.mean(data, axis = 0)
plt.plot(day_avg)
plt.title('Average Inflammation of Days')
plt.xlabel('Day')
plt.ylabel('Average')
plt.show()
````

#### Qestion 3:
Given the following list of file names: 
```
filenames = ['inflammation-02.csv', 'inflammation-04.csv', 'inflammation-06.csv']
```
Write a program to plot the daily averages, maximums, and minimums of inflammation for each of the file. Your plot should have three rows each of which shows side-by-side plots of average, max, and min. The following figure illustrates the expected outcome:
![](assignment2-question3.png)

````python
# Write your code here.
import numpy as np
import matplotlib.pyplot as plt
# imports numpy and pyplot

filenames = ['inflammation-02.csv', 'inflammation-04.csv', 'inflammation-06.csv']
# list of file names

for file in filenames:
# loop over files

    plt.figure(figsize = (20,6))
    # setting plot figure size
    
    plt.subplot(131)
    # creates subplot
    
    data = np.loadtxt("inflammation-data/"+file, delimiter = ",")
    # loads data file
    
    plt.plot(np.mean(data, axis = 0))
    # plots the mean along the day axis
    
    plt.grid(which = "major")
    # adds a grid
    
    plt.title('Average Inflammation of Days of '+file[:-4])
    plt.xlabel('Day')
    plt.ylabel('Average')
    # labels title, x-axis, and y-axis
    
    plt.subplot(132)
    plt.plot(np.max(data, axis = 0))
    plt.grid(which = "major")
    plt.title('Maximum Inflammation of Days of '+file[:-4])
    plt.xlabel('Day')
    plt.ylabel('Maximum')
    # plots max value of each day
    
    plt.subplot(133)
    plt.plot(np.min(data, axis = 0))
    plt.grid(which = "major")
    plt.title('Minimum Inflammation of Days of '+file[:-4])
    plt.xlabel('Day')
    plt.ylabel('Minimum')
    # plots min value of each day
    
    plt.show()
    # displays plot
````

#### Question 4: 
Download the data file "winemag-data-130k-v2.csv" and load it into a Pandas DataFrame using the column 0 as index. This data set contains 130k reviews for different wine verieties. Answer the following questions:
* What features do the reviews use?
* What is the value of the "description" for the third view?
* Select the country, province, region_1, and region_2 columns of the records with the 0, 1, 10, and 100 index positions.
* Select all the reviews for wines made in 'Italy'. How many reviews are there for wines made in 'Italy'?
* Suppose the 'points' column represents the rating of a wine. An above-averagely good wine has a value in 'points' equal or greater than 90. Between France and Italy, which country produces more above-averagely good wines?

````python
# Write your code here.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#imports libraries

df = pd.read_csv("winemag-data-130k-v2.csv", index_col=0)
# reads data and sets to variable df

for i in list(df): print("Reviews Include: %s" %i)
# loop over heads of dataframe data set and prints

print("\nVlaue of third row of 'description' column is: %s" %df["description"].iloc[3])
# prints the third description column

print("\n", df[["country", "province", "region_1", "region_2"]].iloc[[0, 1, 10, 100]])
# prints 0, 1, 10, and 100 indexed rows of dataframe data set

print("\nNumber of reviews made in Italy: %s" %df["country"].value_counts().loc["Italy"])
# prints number of reviews from Italy

reviews = df[df["points"] > 90][["country", "points"]].groupby("country")["points"].count().loc[["France", "Italy"]]

print("\nNumber of points > 90 made in France: %s\nNumber of points > 90 made in Italy: %s" %(reviews[0], reviews[1]))
# prints count of better reviewed wines between France and Italy
````

#### Question 5:
Continue to work on the DataFrame of wine review loaded in Question 4. Answer the following questions:
* What countries are represented in the data set?
* What countries appear in the dataset most often?
* What is the median of the price?
* Create a new column 'price_norm' by subtracting the median price from the price column. Use the DataFrame.apply method.
* Create a new column 'bargain' which contains the points-to-price ratio. Which wine is the "best bargain", e.g., which wine has the highest points-to-price ratio in the dataset?

````python
# Write your code here.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#imports libraries

df = pd.read_csv("winemag-data-130k-v2.csv", index_col=0)
# reads data and sets to variable df

print(df["country"].value_counts())
# prints count of the number of times each country shows up in the dataframe data set

median = df["price"].median()
print("\nThe median of the price is: %s" %median)
# prints out the calculated median 

df["price_norm"] = df["price"].apply(lambda x: x - median)
print("\n%s" %df["price_norm"])
# prints out the new set column "price_norm"

df["bargain"] = df["points"]/df["price"]
print("\n%s" %df.loc[df["bargain"].idxmax()][["variety", "bargain"]])
# prints out the type of wine with the highest points-to-price ratio
````
